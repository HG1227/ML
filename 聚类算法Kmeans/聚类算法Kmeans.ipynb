```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=500,n_features=2,centers=4,random_state=1)
```

```python
X.shape
```

```python
y.shape
```

```python
fig, ax1 = plt.subplots(1)
ax1.scatter(X[:, 0], X[:, 1],
            marker="o",  # 点的形状
            s=8  # 点的大小
            )
plt.show()
```

```python
color = ['red','pink','orange','gray']
fig, ax1 = plt.subplots(1)
for i in range(4):
    ax1.scatter(X[y==i,0],X[y==i,1],
               marker = 'o',
               s = 8,
               c = color[i]
               )
plt.show()    
```

```python
from sklearn.cluster import KMeans
n_cluster = 3
```

```python
cluster = KMeans(n_clusters=n_cluster, random_state=0).fit(X)
```

## `.labels_` 查看聚类后的标签

```python
y_pred = cluster.labels_
y_pred
```

```python
# 和上面的结果是一样的
pre = cluster.fit_predict(X)
```

```python
pre
```

```python
pre == y_pred
```

```python
# 使用部分数据来训练模型
cluster_smallsub = KMeans(n_clusters=n_cluster,random_state=0).fit(X[:200])
```

```python
y_pred_ = cluster_smallsub.predict(X)
```

```python
y_pred_
```

```python
# 结果不完全一样
y_pred == y_pred_
```

## `.cluster_centers_` 查看每个簇的质心（坐标）

```python
centroid = cluster.cluster_centers_
centroid
```

## `.inertia_` 每个样本点到距离它们最近的簇心的均方距离，又叫簇内平方和

```python
inertia = cluster.inertia_
inertia
```

```python
color = ['red', 'pink', 'orange', 'gray']

fig, ax1 = plt.subplots(1)

# 画聚类的结果
for i in range(n_cluster):
    ax1.scatter(X[y_pred == i, 0],
                X[y_pred == i, 1],
                marker='o',
                c=color[i])

# 画每个簇的质心
ax1.scatter(centroid[:, 0], centroid[:, 1],
            marker="x",
            s=15,
            c='black')
# ax1.axis('off')
plt.show()
```

分成 不同的簇，查看簇内平方和的结果

```python
n_cluster = 4
cluster_ = KMeans(n_clusters=n_cluster,random_state=0).fit(X)
inertia = cluster_.inertia_
inertia
```

```python
n_clusters = 5
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
inertia_ = cluster_.inertia_
inertia_
```

```python
n_clusters = 6
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
inertia_ = cluster_.inertia_
inertia_
```

```python
n_clusters = 500
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
inertia_ = cluster_.inertia_
inertia_
```

簇内平方和 不是评价聚类结果好坏的一个指标

# 标签未知的情况，利用轮廓系数，来评价聚类的结果

模块`metrics`中的类`silhouette_score`来计算轮廓系数,它返回的是一个数据集中，所有样本的轮廓系数的均值

`metrics`模块中的`silhouette_sample`，它的参数与轮廓系数一致，但返回的
是数据集中每个样本自己的轮廓系数。


```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
```

```python
X.shape
```

```python
y_pred
```

```python
# 分 3 簇的情况下
silhouette_score(X,y_pred)
```

```python
n_cluster = 5
cluster_ = KMeans(n_clusters=n_cluster,random_state=0).fit(X)
```

```python
# 分 4 簇
silhouette_score(X,cluster_.labels_)
```

```python
# 分 5 簇
silhouette_score(X,cluster_.labels_) #分5簇
```

查看 `silhouette_sample` 的结果

```python
# 3 簇的情况
silhouette_samples(X,y_pred).mean()
```

# 当真实标签未知的时候：`Calinski-Harabaz Index`

卡林斯基-哈拉巴斯指数（Calinski-Harabaz Index，简称CHI，也被称为方差
比标准）

```python
from sklearn.metrics import calinski_harabaz_score
```

```python
calinski_harabaz_score(X, y_pred)
```
